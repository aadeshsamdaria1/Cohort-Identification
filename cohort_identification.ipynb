2.***Clinicians utilize various criteria to characterize hypotension:***

Undifferentiated Shock (may include hypotension):

Serum Lactate Level Elevation: An elevation of serum lactate levels exceeding 4 mmol/L.

Absolute Hypotension:

* Systolic blood pressure below 90 mmHg.
* Mean arterial pressure below 65 mmHg.

Relative Hypotension: A drop in systolic blood pressure by more than 40 mmHg.

Orthostatic Hypotension: Manifests as:

* A fall of more than 20 mmHg in systolic pressure.
* A fall of more than 10 mmHg in diastolic pressure upon standing.

Profound Hypotension: Characterized by dependence on vasopressors for support.

***Overall Hypotension Conditions:***

Hypotension is generally identified through the following conditions:

* Serum lactate levels increasing by 4 mmol/L or more.

* Systolic blood pressure below 90 mmHg or diastolic blood pressure below 60 mmHg (while lying or sitting).
* Mean arterial pressure below 65 mmHg (while lying, sitting, or standing).
* Systolic blood pressure below 70 mmHg or diastolic blood pressure below 50 mmHg (while standing).
* Profound hypotension, which necessitates the use of vasopressors for management.

The below sql code fetches the hypotension data from the mimic tables. The main set of data is taken from the chartsevents table. The filters are the itemid must be one of 220050, 220179, 224167, 227243, 220051, 220180, 224643, 227242, 220052, 220181, which represent the blood pressure values. The bp values must lie between 0 and 300. The drugs administered are taken from the emar table and only the drugs that have an event_txt confirmed, applied, or administered are selected. ICD codes are also fetched from the diagnoses_icd table. The data is grouped together based on subject_id, hadm_id, and itemid to reduce the number of rows. Data can be printed by uncommenting the df line.

```python
# Query to fetch hypotension related data from MIMIC-IV
df = run_query("""
WITH HypotensionData AS (
  SELECT ce.subject_id,
    ce.hadm_id,
    ce.itemid,
    STRING_AGG(CAST(emar.charttime AS STRING), ',') AS Drug_administered_times,
    STRING_AGG(CAST(ce.charttime AS STRING), ',') AS Diagonsed_times,
    STRING_AGG(CAST(ce.valuenum AS STRING), ',') AS values,
    STRING_AGG(CAST(emar.medication AS STRING), ',') AS Medications,
    STRING_AGG(CAST(emar.event_txt AS STRING), ',') AS Status,
FROM
    `physionet-data.mimiciv_icu.chartevents` AS ce
INNER JOIN
    `physionet-data.mimiciv_hosp.emar` AS emar
    ON emar.subject_id = ce.subject_id AND emar.hadm_id = ce.hadm_id
WHERE
    (ce.valuenum <300 AND ce.valuenum >=0)
    AND (ce.itemid IN (220050, 220179, 224167, 227243, 220051, 220180, 224643, 227242, 220052, 220181))
    AND emar.event_txt IN ('Confirmed', 'Administered', 'Applied')
    AND (
        emar.medication LIKE '%Vasopressin%'
        OR emar.medication LIKE '%Dopamine%'
        OR emar.medication LIKE '%Dobutamine%'
        OR emar.medication LIKE '%Adrenaline%'
        OR emar.medication LIKE '%Noradrenaline%'
        OR emar.medication LIKE '%Isoprenaline%'
        OR emar.medication LIKE '%Dopexamine%'
        OR emar.medication LIKE '%Milrinone%'
        OR emar.medication LIKE '%Amrinone%'
    )
    GROUP BY
      ce.subject_id,
      ce.hadm_id,
      ce.itemid
)
SELECT
    hd.subject_id,
    hd.hadm_id,
    hd.itemid,
    hd.Diagonsed_times,
    hd.Drug_administered_times,
    hd.values,
    hd.Medications,
    hd.Status,
    STRING_AGG(CAST(dicd.icd_code AS STRING),',') AS ICD_Codes,
    STRING_AGG(CAST(dicd.icd_version AS STRING),',') AS ICD_Versions
FROM
    HypotensionData AS hd
LEFT JOIN
    `physionet-data.mimiciv_hosp.diagnoses_icd` AS dicd
ON hd.subject_id = dicd.subject_id AND hd.hadm_id = dicd.hadm_id
GROUP BY
    hd.subject_id,
    hd.hadm_id,
    hd.itemid,
    hd.Diagonsed_times,
    hd.Drug_administered_times,
    hd.values,
    hd.Medications,
    hd.Status;
""")
#df
```

In the following code, the final dataset is obtained by filtering the data from the sql. The threshold is selected based on the item id. It is 90 for systolic blood pressure. 60 for diastolic blood pressure and 65 for mean arterial blood pressure. Bp values less than the threshold are considered hypotension cases. For these cases, two values are recorded. One less than the threshold and the other greater than the threshold. The time difference between these two records must be more than 60 minutes. Also, the ICD code for each row is filtered to only display the ICD codes that match those of Hypotension.

```python
filtered_rows = []
processed_combinations = set()
# Create new empty columns for start and end values
df['start_time'] = None
df['start_value'] = None
df['end_time'] = None
df['end_value'] = None
df['start_drug_administered_time'] = None
hypotension_codes=['45821','4589','I95','I952','I953','I959'] # Hypotensions codes from d_icd_diagnoses
# SQL code to get get hypotension codes
# SELECT * FROM `physionet-data.mimiciv_hosp.d_icd_diagnoses` where long_title like '%Hypotension%' LIMIT 1000

for index, row in df.iterrows():
  if row['subject_id'] in processed_combinations: # skip row if parient already present in cohort
        continue
  try:
    time_frames = [datetime.strptime(ts, '%Y-%m-%d %H:%M:%S') for ts in row['Diagonsed_times'].split(',')]
    blood_pressure_values = [int(bp) for bp in row['values'].split(',')]
    if row['itemid'] in [220050, 220179, 224167, 227243]:  # Systolic blood pressure less than 90 is considered hypotension
      bp_limit = 90
    elif row['itemid'] in [220051, 220180, 224643, 227242]:  # Diastolic blood pressure less than 60 is considered hypotension
      bp_limit = 60
    elif row['itemid'] in [220052, 220181]: # Mean Arterial blood pressure less than 65 is considered hypotension
      bp_limit = 65
    else:
      break

    start_index = None
    end_index = None
    time_diff = None
    # Find the start and end indexes
    for i, bp in enumerate(blood_pressure_values):
        if bp < bp_limit and start_index is None:
            start_index = i
        elif bp >= bp_limit and start_index is not None:
            end_index = i
            time_diff = time_frames[end_index] - time_frames[start_index]
            if time_diff >= timedelta(minutes=60):
                break
            else:
                start_index = None
                end_index = None
                time_diff = None
    first_time_drug_administered = sorted([datetime.strptime(ts, '%Y-%m-%d %H:%M:%S') for ts in row['Drug_administered_times'].split(',')])[0]
    if start_index is not None and end_index is not None and time_frames[end_index] > first_time_drug_administered and time_frames[start_index] < first_time_drug_administered:
        filtered_rows.append(index)
        processed_combinations.add(row['subject_id'])
        # Update the new columns with start and end values
        df.at[index, 'start_time'] = time_frames[start_index]
        df.at[index, 'start_value'] = blood_pressure_values[start_index]
        df.at[index, 'end_time'] = time_frames[end_index]
        df.at[index, 'end_value'] = blood_pressure_values[end_index]
        # Replace Drug_administered_times with only the first value
        df.at[index, 'start_drug_administered_time'] = first_time_drug_administered
        icd_codes=row['ICD_Codes'].split(',')
        common_elements = np.intersect1d(icd_codes, hypotension_codes)  # Checking to see if any of the hypotention codes are present
        df.at[index, 'ICD_Codes'] = common_elements
  except ValueError:
    continue
# Filter the DataFrame
filtered_df = df.loc[filtered_rows]
# Replace elements in Medication and Status columns with unique values
filtered_df['Medications'] = filtered_df['Medications'].apply(lambda meds: ','.join(set(meds.split(','))))
filtered_df['Status'] = filtered_df['Status'].apply(lambda status: ','.join(set(status.split(','))))

```

```python
with warnings.catch_warnings():
  warnings.simplefilter("ignore", category=DeprecationWarning)
  data = filtered_df[['subject_id','hadm_id','itemid', 'start_time','start_value','end_time','end_value','start_drug_administered_time','Medications','Status', 'ICD_Codes']]
  data = pd.DataFrame(data)
  # Final cohort of patients
  data = data.sort_values(by='ICD_Codes') # Sorted to display the presence of ICD Codes
  data['ICD_Codes'] = data['ICD_Codes'].apply(lambda x: '-' if not x else x)
  data.index = range(1, len(data) + 1)
data
```
